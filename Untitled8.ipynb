```python
import pandas as pd
```

```python
data = {
    "user_id":[1,1,1,1,1,2,2,2,3,3,4,4,4],

    "event":[
        "app_open",
        "restaurant_view",
        "add_to_cart",
        "checkout",
        "order_complete",
        "app_open",
        "restaurant_view",
        "add_to_cart",
        "app_open",
        "restaurant_view",
        "app_open",
        "restaurant_view",
        "add_to_cart"
    ]
}

df = pd.DataFrame(data)

df
```

```python
df.head()
```

```python
df['event'].value_counts()
```

```python
funnel = df.groupby('event')['user_id'].nunique()

funnel
```

```python
funnel = df.groupby('event')['user_id'].nunique()

conversion = funnel / funnel.shift(1)

conversion
```

```python
conversion.sort_values()
```

```python
import matplotlib.pyplot as plt

funnel = df.groupby('event')['user_id'].nunique()

plt.figure()
funnel.plot(kind='bar')

plt.xlabel("Funnel Stage")
plt.ylabel("Number of Users")
plt.title("Zomato Ordering Funnel Analysis")

plt.show()
```
